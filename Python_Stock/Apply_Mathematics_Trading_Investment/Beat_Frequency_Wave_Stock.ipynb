# Beat Frequency Wave Equation

https://en.wikipedia.org/wiki/List_of_equations_in_wave_theory

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define a function for a beat frequency equation (sine wave as an example)
def beat_frequency_wave(t, A1, f1, A2, f2):
    wave1 = A1 * np.sin(2 * np.pi * f1 * t)
    wave2 = A2 * np.sin(2 * np.pi * f2 * t)
    return wave1 + wave2

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values
dates = dataset.index

# Create a time array based on the date index
time = np.arange(len(data))

# Parameters for the beat frequency equation
A1 = 1.0  # Amplitude of the first wave
f1 = 0.02  # Frequency of the first wave
A2 = 0.5  # Amplitude of the second wave
f2 = 0.05  # Frequency of the second wave

# Calculate the beat frequency wave
beat_wave = beat_frequency_wave(time, A1, f1, A2, f2)

# Plot the stock data and the beat frequency wave
plt.figure(figsize=(10, 6))
plt.plot(dates, data, label='Stock Data')
plt.plot(dates, beat_wave, label='Beat Frequency Wave')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Stock Data and Beat Frequency Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot the stock data on the first subplot
ax1.plot(dates, data, label='Stock Data')
ax1.set_xlabel('Date')
ax1.set_ylabel('Value')
ax1.set_title('Stock Data')
ax1.legend()
ax1.grid(True)

# Plot the beat frequency wave on the second subplot
ax2.plot(dates, beat_wave, label='Beat Frequency Wave', color='orange')
ax2.set_xlabel('Date')
ax2.set_ylabel('Value')
ax2.set_title('Beat Frequency Wave')
ax2.legend()
ax2.grid(True)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```
