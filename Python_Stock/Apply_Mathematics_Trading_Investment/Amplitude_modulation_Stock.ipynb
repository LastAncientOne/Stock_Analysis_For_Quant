# Amplitude modulation

https://en.wikipedia.org/wiki/Amplitude_modulation

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values

# Define the Amplitude Modulation (AM) function
def amplitude_modulation(signal, carrier_frequency, modulation_index):
    # Create a time array
    time = np.arange(0, len(signal)) / len(signal)
    
    # Create the carrier signal
    carrier_signal = np.cos(2 * np.pi * carrier_frequency * time)
    
    # Modulate the signal
    modulated_signal = (1 + modulation_index * signal) * carrier_signal
    
    return modulated_signal

# Set the carrier frequency and modulation index
carrier_frequency = 0.1  # Adjust as needed
modulation_index = 0.5   # Adjust as needed

# Apply AM to the stock data
amplitude_modulated_data = amplitude_modulation(data, carrier_frequency, modulation_index)

# Plot the original and AM-modulated signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(data, label='Original Signal')
plt.title('Original Stock Price Data')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(amplitude_modulated_data, label='AM Modulated Signal', color='red')
plt.title('AM Modulated Stock Price Data')
plt.legend()

plt.tight_layout()
plt.show()
```
