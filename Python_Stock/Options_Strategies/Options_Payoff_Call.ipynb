# Payoff for a Call Option

```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
dfo = yf.Ticker("AAPL")
```

```python
dfo.options
```

```python
dfo_exp = dfo.option_chain('2020-04-16')
```

```python
dfo_exp.calls
```

```python
df = yf.download("AAPL")
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Adj Close'][-1]
```

```python
dfo_exp.calls['strike'][31]
```

```python
def payoff_call(St, X):
    poc = (St-X+abs(St-X)) / 2
    return poc
```

```python
St = df['Adj Close'][-1] # Current Price
X = dfo_exp.calls['strike'][31] # Strike Price
```

```python
payoff_call(St, X)
```
