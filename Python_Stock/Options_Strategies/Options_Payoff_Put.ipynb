# Payoff for a Put Option

```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
dfo = yf.Ticker("AAPL")
```

```python
dfo.options
```

```python
dfo_exp = dfo.option_chain('2020-04-16')
```

```python
dfo_exp.puts
```

```python
df = yf.download("AAPL")
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Adj Close'][-1]
```

```python
dfo_exp.puts['strike'][29]
```

```python
def payoff_puts(St, X):
    pop = (X-St+abs(X-St)) / 2
    return pop
```

```python
St = df['Adj Close'][-1] # Current Price
X = dfo_exp.puts['strike'][29] # Strike Price
```

```python
payoff_puts(St, X)
```

```python
St = df['Adj Close'][-1] # Current Price
X = dfo_exp.puts['strike'][30] # Strike Price
```

```python
payoff_puts(St, X)
```
